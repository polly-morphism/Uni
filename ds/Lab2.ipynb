```python
import pandas as pd
from scipy.stats import skew, kurtosis
data=pd.read_csv('A2.txt', header=None)
```

```python
statistics=pd.DataFrame()
statistics['середнє']=data.mean(axis=0)
statistics['дисперсія']=data.var(axis=0)
statistics['медіана']=data.median(axis=0)
statistics['коефіцієнт ексцесу']=kurtosis(data, axis=0)
statistics['коефіцієнт асиметрії']=skew(data, axis=0)
statistics
```

```python

data=(data-data.mean(axis=0))/(data.std(axis=0))
data.head()
```

```python
corr_matrix=data.corr()
print("Матриця кореляції: ")
corr_matrix
```

```python
corr_pairs=corr_matrix.unstack().sort_values()
strong_corr=corr_pairs[abs(corr_pairs)>0.9]
lst=[strong_corr.index[i] for i in range(0, strong_corr.shape[0]-12, 2)]
print(strong_corr[lst])
a=3
b=1
c=11
d=10
print("Обрані параметри з великою парною кореляцією: a={}, b={}, c={}, d={}".format(a, b, c, d))
```

```python
#Відповідні попарні коефіцієнти кореляції
r_ab=strong_corr[a,b]
r_ac=strong_corr[a,c]
r_ad=strong_corr[a,d]
r_bc=strong_corr[b,c]
r_bd=strong_corr[b,d]
r_cd=strong_corr[c,d]
print("r_ab={}, r_ac={}, r_ad={}\n\nr_bc={}, r_bd={}, r_cd={}".format(r_ab, r_ac, r_ad, r_bc, r_bd, r_cd))
```

```python
r_ab_c=(r_ab-r_ac*r_bc)/(((1-r_ac**2)*(1-r_bc**2))**0.5)
r_ac_b=(r_ac-r_ab*r_bc)/(((1-r_ab**2)*(1-r_bc**2))**0.5)
print("Частковий коефіцієнт кореляції між ознаками a  та b без урахування впливу ознаки c: ", r_ab_c)
print("Частковий коефіцієнт кореляції між ознаками a  та c без урахування впливу ознаки b: ", r_ac_b)
```

```python
r_ad_c=(r_ad-r_ac*r_cd)/(((1-r_ac**2)*(1-r_cd**2))**0.5)
r_bd_c=(r_bd-r_bc*r_cd)/(((1-r_bc**2)*(1-r_cd**2))**0.5)
print("Частковий коефіцієнт кореляції між ознаками a  та d без урахування впливу ознаки c: ", r_ad_c)
print("Частковий коефіцієнт кореляції між ознаками d  та b без урахування впливу ознаки c: ", r_bd_c)
r_ab_cd=(r_ab_c-r_ad_c*r_bd_c)/(((1-r_ad_c**2)*(1-r_bd_c**2))**0.5)
print("Частковий коефіцієнт кореляції між ознаками a  та b без урахування впливу ознаки c та d: ", r_ab_cd)
```

```python
r_ad_b=(r_ad-r_ab*r_bd)/(((1-r_ab**2)*(1-r_bd**2))**0.5)
r_cd_b=(r_cd-r_bc*r_bd)/(((1-r_bc**2)*(1-r_bd**2))**0.5)
print("Частковий коефіцієнт кореляції між ознаками a  та d без урахування впливу ознаки b: ", r_ad_b)
print("Частковий коефіцієнт кореляції між ознаками c  та d без урахування впливу ознаки b: ", r_cd_b)
r_ac_bd=(r_ac_b-r_ad_b*r_cd_b)/(((1-r_ad_b**2)*(1-r_cd_b**2))**0.5)
print("Частковий коефіцієнт кореляції між ознаками a  та c без урахування впливу ознаки b та d: ", r_ac_bd)
```

```python
r_ad_bc=(r_ad_b-r_ac_b*r_cd_b)/(((1-r_ac_b**2)*(1-r_cd_b**2))**0.5)
print("Частковий коефіцієнт кореляції між ознаками a  та d без урахування впливу ознаки b та c: ", r_ad_bc)
```
