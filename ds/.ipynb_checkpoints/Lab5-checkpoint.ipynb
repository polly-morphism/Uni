## Алгоритм 1. Класичний метод.

```python
def distance(alpha, beta, metric):
    res=[]
    if metric=='E':
        n=2
    elif metric=='M':
        n=int(input("Введіть r для для обчислення:"))
    elif metric=="H":
        n=1
    if metric !='C':
        for m in beta:
            d=0
            for i in range(7):
                temp=0
                for j in range(11):
                    if alpha[j][i]!=m[j][i]:temp+=1
                d+=temp**n
            res.append(pow(d,1/n))
    else:
        for m in beta:
            d=[]
            for i in range(7):
                temp=0
                for j in range(11):
                    if alpha[j][i]!=m[j][i]:temp+=1
                d.append(temp)
            res.append(max(d))
    return res
```

## Алгоритм 2. Нейронна мережа Хопфілда.
Перероблюємо всі образи з формату матриць у формат векторів:

```python
def vector(M):
    temp=[]
    for i in range(11):
        for j in range(7):
            if M[i][j]==0:
                temp.append(-1)
            else: 
                temp.append(1)
    return temp
ML=list(map(vector,ML))
X=vector(X)
```

Створюємо та навчаємо мережу Хопфілда:

```python
W=np.zeros((77, 77))
for m in ML:
    temp=np.array([m]).transpose().dot(np.array([m]))
    W+=temp
for i in range(77):
    W[i][i]=0
W/=77
```

Тепер передаємо Нейроній мережі образ Х для розпізнавання: 

```python
Z=[]
Z=np.sign(np.dot(W,np.array([X]).transpose()))
print(Z.reshape(11,7)==np.array(ML[9]).reshape(11,7))
Z.reshape(11,7)
```

Наш образ Х нейроная мережа Хопфілда розпізнала з першого разу як образ числа 9
## Завдання 2. Провести кластеризацію даних із DataSet

### A)

```python
df = pd.read_csv('A6.txt', header=None, names=[f"Parameter {x}" for x in range(1,13)])
df
```

```python
k1=5
k2=7
X1 = df.values[::]
X1 = np.nan_to_num(X1)
kmeans1 = KMeans(init = "k-means++", n_clusters=k1)
y1=kmeans1.fit_predict(X1)
kmeans2 = KMeans(init = "k-means++", n_clusters=k2)
y2=kmeans2.fit_predict(X1)
```

Кластерізація для к = 5 та розмірності точок 12:

```python
df['cluster']=y1
df.groupby("cluster").count()
```

Кластерізація для к = 7 та розмірності точок 12:

```python
df['cluster']=y2
df.groupby("cluster").count()
```

### Б) Кластеризація даних трьох головних факторів

```python
X2=df[['Parameter 1','Parameter 2','Parameter 3']]
kmeans11 = KMeans(init = "k-means++", n_clusters=k1)
y11=kmeans1.fit_predict(X2)
kmeans21 = KMeans(init = "k-means++", n_clusters=k2)
y21=kmeans21.fit_predict(X2)
```

Кластерізація для к = 5 та розмірності точок 3:

```python
df['cluster']=y11
df.groupby("cluster").count()
```

```python
df['cluster']=y21
df.groupby("cluster").count()
```
