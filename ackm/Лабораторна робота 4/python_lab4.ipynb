```python
import numpy as np
import math


X = np.array([-1.0, -2.0, -3.0, -4.0])
true_min = np.array([1.0, 1.0, 1.0, 1.0])
learning_rate = 0.0001
eps = 0.00001
max_iter = 100000


def objective_function(x):
    return (
        (x[0] - 1) ** 2
        + 10 * (x[1] - 1) ** 2
        + 100 * (x[2] - 1) ** 2
        + 1000 * (x[3] - 1) ** 2
    )


def gradient(x):
    d_x0 = 2 * (x[0] - 1)
    d_x1 = 20 * (x[1] - 1)
    d_x2 = 200 * (x[2] - 1)
    d_x3 = 2000 * (x[3] - 1)
    return np.array([d_x0, d_x1, d_x2, d_x3], dtype=np.float32)


def minimize(x):
    error = 1
    itr = 0
    curr_x = x
    err_list = []
    while error > eps or itr < max_iter:
        prev_x = curr_x
        curr_x = curr_x - learning_rate * gradient(prev_x)
        error = abs(objective_function(curr_x) - objective_function(prev_x))
        err_list.append(error)
        itr += 1
    return curr_x, err_list


solution, errors = minimize(X)

print("minimum:", solution)
print("minimum value:", objective_function(solution))
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20,10))
plt.plot(range(len(errors)), errors, "r", label="error")
plt.legend()
plt.show()
```
