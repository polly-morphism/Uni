```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')

def Runge_Kutta(x0,y0,t0,tn,h,f1,f2):

    x,y = x0,y0
    X,Y = [x], [y]

    for t in np.arange(t0,tn,h):

        q1 = h*f1(x,y)
        k1 = h*f2(x,y)

        q2 = h*f1(x+(q1/2),y+(k1/2))
        k2 = h*f2(x+(q1/2),y+(k1/2))

        q3 = h*f1(x+(q2/2),y+(k2/2))
        k3 = h*f2(x+(q2/2),y+(k2/2))

        q4 = h*f1(x+q3,y+k3)
        k4 = h*f2(x+q3,y+k3)

        x = x + (1/6)*(q1 + 2 * q2 + 2 * q3 + q4)
        y = y + (1/6)*(k1 + 2 * k2 + 2 * k3 + k4)

        X.append(x)
        Y.append(y)

    return X,Y

def Lotka_Volterra(a,b,c,d,x0, y0, t0, tn,epsilon):
    h = epsilon ** (1/4)
    n = ceil((tn-t0)/h)

    if n % 2:
        n+=1
    h = (tn-t0)/n

    def f1(x,y): return a*x - b*x*y
    def f2(x,y): return -c*y + d*x*y

    X,Y = Runge_Kutta(x0,y0,t0,tn,h,f1,f2)	

    while True:
        h /= 2
        X2, Y2 = Runge_Kutta(x0,y0,t0,tn,h,f1,f2)
        r1 = max([abs(x - x2) for x, x2 in zip(X, X2[::2])])
        r2 = max([abs(y - y2) for y, y2 in zip(Y, Y2[::2])])
        r = max(r1, r2)
        X, Y = X2, Y2
        if r < epsilon:
            return X2, Y2, h, c / d, a / b


epsilon = 0.001
a,b,c,d = 1.28, 1.09, 0.87, 0.35
x0,y0 = 2.55 , 1.56
t0,tn = 2, 32

plt.style.use('dark_background')
X2, Y2, h, xkr,ykr = Lotka_Volterra(a,b,c,d,x0, y0, t0, tn,epsilon)
T = np.arange(t0, tn + h / 2, h)
plt.title("Графіки популяцій з часом")
plt.plot(T, X2, label = 'популяція здобичі')
plt.plot(T, Y2, label = 'популяція хижака')
plt.legend(loc= 'upper right')
plt.show()

fig=plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')

plt.title("Фазовий портрет популяцій")
plt.scatter(x0,y0, label = 'початковий стан', color='red')
plt.scatter(xkr,ykr, label = 'критична точка')

plt.plot(X2,Y2,label = 'фазова крива')
plt.legend(loc= 'upper right')
plt.show()
```

```python
from math import *
import numpy as np
import matplotlib.pyplot as plt


def f1(t, x, y):
    return a * x - b * x * y


def f2(t, x, y):
    return -c * y + d * x * y


def Runge_Kutta(x, y, start, finish, h):
    X, Y = [x], [y]

    for t in np.arange(start, finish, h):  # Count k_i, q_i formula in pdf

        q1 = h * f1(t, x, y)
        k1 = h * f2(t, x, y)

        q2 = h * f1(t, x + (q1 / 2), y + (k1 / 2))
        k2 = h * f2(t, x + (q1 / 2), y + (k1 / 2))

        q3 = h * f1(t, x + (q2 / 2), y + (k2 / 2))
        k3 = h * f2(t, x + (q2 / 2), y + (k2 / 2))

        q4 = h * f1(t, x + q3, y + k3)
        k4 = h * f2(t, x + q3, y + k3)

        x = x + (1 / 6) * (q1 + 2 * q2 + 2 * q3 + q4)
        y = y + (1 / 6) * (k1 + 2 * k2 + 2 * k3 + k4)

        X.append(x)
        Y.append(y)

    return X, Y


def Lotka_Volterra(a, b, c, d, x0, y0, start, finish, epsilon):
    r = 0
    h = np.sqrt(np.sqrt(epsilon))  # Length of step
    n = ceil((finish - start) / h)  # Number of steps

    h = (finish - start) / n

    X, Y = Runge_Kutta(x0, y0, start, finish, h)

    while r < epsilon:
        h /= 2
        X2, Y2 = Runge_Kutta(x0, y0, start, finish, h)
        r1 = max([abs((x - x2) / 15) for x, x2 in zip(X, X2[::2])])
        r2 = max([abs((y - y2) / 15) for y, y2 in zip(Y, Y2[::2])])
        r = max(r1, r2)
        X, Y = X2, Y2
    return X2, Y2, h, c / d, a / b


epsilon = 0.001
a,b,c,d = 1.28, 1.09, 0.87, 0.35
x0 = 2.55
y0 = 1.56
start = 2
finish = 32


plt.style.use('dark_background')

fig, axes = plt.subplots(1, 2, subplot_kw=dict())
X2, Y2, h, xkr, ykr = Lotka_Volterra(a, b, c, d, x0, y0, start, finish, epsilon)
T = np.arange(start, finish + h / 2, h)

axes[0].plot(T, X2, "c", label="Популяція здобичі")
axes[0].plot(T, Y2, "m", label="Популяція хижака")
axes[0].legend(loc="upper right")
axes[1].scatter(x0, y0, label="Початковий стан")
axes[1].scatter(xkr, ykr, label="Критична точка")

axes[1].plot(X2, Y2, "g", label="Фазова крива")
axes[1].legend(loc="upper right")
axes[0].grid(True)
axes[1].grid(True)
axes[0].legend()
axes[1].legend()
fig=plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')

plt.show()
```
